```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('4_6.csv')
mean_score = df['mean_score'].values
max_score = df['max_score'].values
min_score = df['min_score'].values
episodes = []
for i in range(1,df.shape[0]+1):
    episodes.append(1000*i)
plt.plot(episodes, mean_score, label = "mean_score")
plt.plot(episodes, max_score, label = "max_score")
plt.plot(episodes, min_score, label = "min_score")
plt.title('Scores')
plt.xlabel('Episodes')
plt.ylabel('Scores')
plt.legend()
plt.show()
```
