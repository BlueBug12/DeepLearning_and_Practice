```python
import pickle
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(fname,title):
    with open(fname, 'rb') as f:
        record_dict = pickle.load(f)
    epoch = record_dict['epoch']
    kld = record_dict['kld']/600
    mse = record_dict['mse']
    beta = record_dict['beta']
    tf_ratio = record_dict['tf_ratio']
    loss = record_dict['loss']
    psnr = record_dict['psnr']
    plt.title(title)
    #plt.plot(epoch,kld,label="kld")    
    #plt.plot(epoch,mse,label="mse")
    #plt.plot(epoch,beta,label="beta")
    #plt.plot(epoch,loss,label="loss")    
    plt.plot(epoch,beta,label="beta")


    plt.xlabel('Epoch')
    plt.ylabel('loss')
    plt.legend()
    plt.show()
```

```python
def plot2(fname,d1,d2,title):
    with open(fname, 'rb') as f:
        record_dict = pickle.load(f)
    epoch = record_dict['epoch']
    data1 = record_dict[d1]
    data2 = record_dict[d2]
    fig,ax = plt.subplots(dpi=150)
    ax.set_title(title)
    ax.plot(epoch, data1, color="red", marker=",")
    ax.set_xlabel("epoch",fontsize=14)
    ax.set_ylabel(d1,color="red",fontsize=14)
    ax2=ax.twinx()
    
    ax2.plot(epoch, data2,color="blue",marker=",")
    ax2.set_ylabel(d2,color="blue",fontsize=14)
    plt.show()
    '''
    fig.savefig('two_different_y_axis_for_single_python_plot_with_twinx.jpg',
                format='jpeg',
                dpi=100,
                bbox_inches='tight')
    '''
    
```

```python
def comp_plot2(f1,f2,data,title,beg,end):
    with open(f1, 'rb') as f:
        record_dict1 = pickle.load(f)
    with open(f2, 'rb') as f:
        record_dict2 = pickle.load(f)
    epoch = record_dict1['epoch'][beg:end]
    d1 = record_dict1[data][beg:end]    
    d2 = record_dict2[data][beg:end]
    plt.figure(dpi=150)
    plt.title(title)
    #plt.plot(epoch,kld,label="kld")    
    #plt.plot(epoch,mse,label="mse")
    #plt.plot(epoch,beta,label="beta")
    #plt.plot(epoch,loss,label="loss")    
    plt.plot(epoch,d1,label="lr=0.001")
    plt.plot(epoch,d2,label="lr=0.01")



    plt.xlabel('Epoch')
    plt.ylabel(data)
    plt.legend()
    plt.show()
```

```python
plot2("./logs/final/cVAE_radam/record.pickle","psnr","loss","Psnr V.S. Loss")
```

```python
comp_plot2("./logs/final/cVAE_radam/record.pickle","./logs/final/cVAE_radam_2/record.pickle","psnr","Learning Rate Comparison",0,300)
```
